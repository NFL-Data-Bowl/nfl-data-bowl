```python
import pandas as pd
```

```python
df=pd.read_parquet('all_play.parquet')
```

```python
ef= pd.read_parquet('scoring_system.parquet')
```

```python
df
```

```python
ef
```

```python
df=df.merge(ef,how="left", on=["gameId","playId"])
```

```python
df
```

```python
df.fillna(0,inplace=True)
```

```python
X = df.drop(columns=["isConfusedPlay", "gameId", "playId"])  # Remove target and non-relevant IDs
y = df["isConfusedPlay"]  # Convert target to integer class labels
```

```python

from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X, y = sm.fit_resample(X, y)
```

```python
X.shape, y.shape
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
sc=StandardScaler()
from sklearn.decomposition import PCA
X_pca= sc.fit_transform(X)
pca = PCA(n_components=500)
principal_components=pca.fit_transform(X_pca)
```

```python
principal_components.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Define features and target
#X = df.drop(columns=["isConfused", "gameId", "playId"])  # Remove target and non-relevant IDs
#y = df["isConfusedPlay"]  # Convert target to integer class labels

# Split data
X_train, X_test, y_train, y_test = train_test_split(principal_components, y, test_size=0.2, random_state=42, stratify=y)


# Train SVC model
svc_model = SVC(kernel="rbf", C=1.0, random_state=42)
svc_model.fit(X_train,y_train)

# Predictions
y_pred = svc_model.predict(X_test)

# Evaluate model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

accuracy, report
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Explained variance
plt.figure(figsize=(10, 5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
auc = roc_auc_score(y_test, y_pred)
auc
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(
    principal_components[:, 0],  # PC1
    principal_components[:, 1],  # PC2
    principal_components[:, 2],  # PC3
    c=y, cmap='viridis', alpha=0.6
)

ax.set_xlabel('PCA Component 1')
ax.set_ylabel('PCA Component 2')
ax.set_zlabel('PCA Component 3')
ax.set_title('3D PCA Projection')

legend = ax.legend(*scatter.legend_elements(), title="Class")
ax.add_artist(legend)

plt.show()
```
