```python
import pandas as pd

plays = pd.read_csv("~/nfl-data-bowl/Sample_Data/Raw/plays.csv")

motion_feature = pd.read_csv("~/nfl-data-bowl/Sample_Data/Raw/processed_offensive_motion_plays.csv")
motion_feature = motion_feature.rename(columns={'playID': 'playId'})
motion_feature = motion_feature.rename(columns={'gameID': 'gameId'})
```

```python
motion_feature.describe()
```

```python
plays_e = plays[["gameId", "playId","expectedPointsAdded","yardsGained"]]

# Join each distance group with plays
motion_merged = plays_e.merge(motion_feature, on=['gameId', 'playId'], how='inner')

motion_merged['bigYardPlay'] = (motion_merged['yardsGained'] > 15).astype(int)
motion_merged['bigEpaPlay'] = (motion_merged['expectedPointsAdded'] > 1.5).astype(int)

print(motion_merged.head)
```

##### Check plays with no offensive motion vs plays with offensive motion

###### Higher EPA and Average Yards for plays where offense go in motion. On average, there are more motion plays with an EPA > 1.5 or yards > 15.

```python
no_motion = motion_merged[motion_merged["num_motions"] == 0]
motion = motion_merged[motion_merged["num_motions"] != 0]

no_motion.describe()
```

```python
motion.describe()
```

```python
print(motion['bigYardPlay'].value_counts())
print(no_motion['bigYardPlay'].value_counts())
print(motion['bigEpaPlay'].value_counts())
print(no_motion['bigEpaPlay'].value_counts())
```

##### Verifying seg_switch_count

###### higher seg_switch_count results in higher yards and epa on average and more big plays

```python
q1_dist = motion['segment_switch_count'].quantile(0.25)
q3_dist = motion['segment_switch_count'].quantile(0.75)

dist_low = motion[motion['segment_switch_count'] <= q1_dist]
dist_mid = motion[(motion['segment_switch_count'] > q1_dist) & 
                        (motion['segment_switch_count'] <= q3_dist)]
dist_high = motion[motion['segment_switch_count'] > q3_dist]

def print_group_stats(group, group_name):
    epa_mean = group["expectedPointsAdded"].mean()
    yards_mean = group['yardsGained'].mean()
    print(f"{group_name} — EPA Mean: {epa_mean:.3f}, Yards Gained Mean: {yards_mean:.2f}")

print_group_stats(dist_low, "Num segs < 25%")
print_group_stats(dist_mid, "Num segs 25%-75%")
print_group_stats(dist_high, "Num segs > 75%")


top_20 = motion.nlargest(100, 'segment_switch_count')
bottom_50 = motion.nsmallest(100, 'segment_switch_count')

print_group_stats(bottom_50 , "bottom100")
print_group_stats(top_20, "top100")
```

```python
print(dist_low['bigYardPlay'].value_counts()) #0.1470
print(dist_mid['bigYardPlay'].value_counts()) #0.1405
print(dist_high['bigYardPlay'].value_counts()) #0.1769
```

#### Verifying num_motions

###### The more motions, the higher the epa and yardage on average.

```python
variable = 'num_motions'

q1_dist = motion[variable].quantile(0.25)
q3_dist = motion[variable].quantile(0.75)

dist_low = motion[motion[variable] <= q1_dist]
dist_mid = motion[(motion[variable] > q1_dist) & 
                        (motion[variable] <= q3_dist)]
dist_high = motion[motion[variable] > q3_dist]

def print_group_stats(group, group_name):
    epa_mean = group["expectedPointsAdded"].mean()
    yards_mean = group['yardsGained'].mean()
    big_y_plays = group['bigYardPlay'].mean()
    big_epa_plays = group['bigEpaPlay'].mean()
    print(f"{group_name} — EPA Mean: {epa_mean:.3f}, Yards Gained Mean: {yards_mean:.2f}, BigYards: {big_y_plays:.2f}, BigEPA: {big_epa_plays:.2f}")

print_group_stats(dist_low, "variable < 25%")
print_group_stats(dist_mid, "variable 25%-75%")
print_group_stats(dist_high, "variable > 75%")


top_20 = motion.nlargest(100, variable)
bottom_50 = motion.nsmallest(100, variable)

print_group_stats(bottom_50 , "bottom100")
print_group_stats(top_20, "top100")
```
